```python
new_df = pd.read_csv("/Users/oli/Uni-Stuff/2022/Project/Code/Saved_data/Logs/Date_2022-03-16_(16-46)_log.csv", header=3)
```

```python
def std(p, trials):
    return np.sqrt(p * (1-p) / trials)

```

```python
new_df['std'] = std(new_df['fix'], 100000)
new_df['plus'] = new_df['fix'] + new_df['std']
new_df['minus'] = new_df['fix'] - new_df['std']

new_df.head()
```

```python
fig, ax = plt.subplots(figsize=(6.4, 4.8), dpi=1000)


ax.plot(new_df['prob'], new_df['fix'])
ax.plot(new_df['prob'], new_df['plus'], linestyle='dashed', color='red')
ax.plot(new_df['prob'], new_df['minus'], linestyle='dashed', color='red'),

ax.set(
    title="Probability of fixation $\mu$, against probability of an edge",
    xlabel="Probability of two people being directly connected.",
    ylabel="Probability of fixation, $\mu$",
)
```

```python
fig.savefig("figs/fixation_against_prob")
```
